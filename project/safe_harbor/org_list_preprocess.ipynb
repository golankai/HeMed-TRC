```python
import pandas as pd
import re
```

```python
path = "organizations_list_raw.csv"
df = pd.read_csv(path)
df
```

```python
# Apply preprocessing - strip leading and trailing whitespaces
df = df.applymap(lambda x: x.strip() if type(x) is str else x)  
df = df.applymap(lambda x: x.strip("\""))
df
```

```python
# Add new rows, change ' to " in the name column
new_rows = pd.DataFrame({"name": df["name"].apply(lambda x: re.sub("'", "\"", x))})
df = pd.concat([df, new_rows], axis=0)
df.reset_index(drop=True, inplace=True)
df = df.dropna()
df = df.drop_duplicates()
df
```

```python
# Save the result
df.to_csv("organizations.csv", index=False)
```
